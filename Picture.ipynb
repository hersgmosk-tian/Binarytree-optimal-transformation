```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def main():
    result = pd.read_csv("results.csv")
    result["time_std_alg1"] = np.sqrt(result.loc[:,"time_var_alg1"])
    result["time_std_alg2"] = np.sqrt(result.loc[:,"time_var_alg2"])
    result["distin_abi_alg1"] = result.loc[:,"time_std_alg1"]/result.loc[:,"time_sum_alg1"]
    result["distin_abi_alg2"] = result.loc[:,"time_std_alg2"]/result.loc[:,"time_sum_alg2"] 
    result_cut = result.iloc[:,[0,1,2,3,4,5,6,11,12]]
    return result_cut
    
    
#     print(result_cut)
```

```python
def draw_mean_time(result):
#     sns.set_theme(style="darkgrid")
    plt.figure(figsize=(8, 6))
    sns.set_style('ticks')
    result_sub = result.loc[:,["time_sum_alg1", "time_sum_alg2"]]
    result_sub.index = result.loc[:,'d']
    flatui = [sns.color_palette("Paired")[1], sns.color_palette("Paired")[5]]
    sns.color_palette(flatui)
    sns.set_palette(sns.color_palette(flatui))
    sns.lineplot(data = result_sub, alpha = 0.8, markers = True)
    plt.xlabel("Depth of BTree")
    plt.ylabel("$Time_{mean}$")
    my_x_ticks = np.arange(4, 17, 1)
    plt.xticks(my_x_ticks)
    plt.legend(labels=['Origin','Optimization'],loc='upper left')
    plt.show()
#     ax = sns.lineplot(data=wide_df)
if __name__ == "__main__":
    result = main()
    draw_mean_time(result.iloc[0:13,:])
```

```python
def draw_distinguish(result):
#     sns.set_theme(style="darkgrid")
    plt.figure(figsize=(8, 6))
    sns.set_style('ticks')
    result_sub = result.loc[:,["distin_abi_alg1", "distin_abi_alg2"]]
    result_sub.index = result.loc[:,'d']
    flatui = [sns.color_palette("Paired")[3], sns.color_palette("Paired")[7]]
    sns.color_palette(flatui)
    sns.set_palette(sns.color_palette(flatui))
    sns.lineplot(data = result_sub, alpha = 1, markers = True)
    plt.xlabel("Depth of BTree")
    plt.ylabel("$Time_{std}/Time_{mean}$")
    my_x_ticks = np.arange(6, 25, 1)
    plt.xticks(my_x_ticks)
    plt.legend(labels=['Origin','Optimization'],loc='upper right')
    plt.show()
#     ax = sns.lineplot(data=wide_df)
if __name__ == "__main__":
    result = main()
#     print(result.columns)
    draw_distinguish(result.iloc[2:,:])
```

```python
def draw_max(result):
#     sns.set_theme(style="darkgrid")
    plt.figure(figsize=(8, 6))
    sns.set_style('ticks')
    result_sub = result.loc[:,['max_time_alg1', 'max_time_alg2']]
    result_sub.index = result.loc[:,'d']
    flatui = ["purple","darkgoldenrod"]
    sns.color_palette(flatui)
    sns.set_palette(sns.color_palette(flatui))
    sns.lineplot(data = result_sub, alpha = 0.8, markers = True)
    plt.xlabel("Depth of BTree")
    plt.ylabel("$Time_{max}$")
    my_x_ticks = np.arange(4, 17, 1)
    my_y_ticks = np.arange(0, 0.7, 0.1)
    plt.xticks(my_x_ticks)
    plt.yticks(my_y_ticks)
    plt.legend(labels=['Origin','Optimization'],loc='upper left')
    plt.show()
#     ax = sns.lineplot(data=wide_df)
if __name__ == "__main__":
    result = main()
    draw_max(result.iloc[:13,:])
```
